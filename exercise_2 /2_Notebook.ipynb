# Dead ReLU

Sigmoid suffers from the vanishing gradients problem, ReLU also has it's drawback.

ReLU neurons output zero and have zero derivatives for all negative inputs. So, if the weights in your network always lead to negative inputs into a ReLU neuron, that neuron is effectively not contributing to the network's training. 

This can be improoved on by for example, initializing the biases with a small positivie number.

# 2 Sigmoid


$$\sigma(x) = \frac{1}{1 + e^{-x}}$$


<h3>Task: Implement.</h3> 
Go into <code>exercise_2/code/activation.py</code> and implement the forward pass of the Sigmoid activation function.
Use the $np.exp$ function to for calculating $e(x)$.

Calculating the gradients might be a great practice for you, so try to do it on a piece of paper and check if your solution is correct!

\begin{aligned}
\frac{d}{d x} \sigma(x) & =\frac{d}{d x}\left[\frac{1}{1+e^{-x}}\right] \\
& =\frac{d}{d x}\left(1+\mathrm{e}^{-x}\right)^{-1} \\
& =-\left(1+e^{-x}\right)^{-2}\left(-e^{-x}\right) \\
& =\frac{e^{-x}}{\left(1+e^{-x}\right)^2} \\
\end{aligned}

We could use this result, but we can do some tricks!

\begin{aligned}
& =\frac{1}{1+e^{-x}} \cdot \frac{e^{-x}}{1+e^{-x}} \\
& =\frac{1}{1+e^{-x}} \cdot \frac{\left(1+e^{-x}\right)-1}{1+e^{-x}} \\
& =\frac{1}{1+e^{-x}} \cdot\left(\frac{1+e^{-x}}{1+e^{-x}}-\frac{1}{1+e^{-x}}\right) \\
& =\frac{1}{1+e^{-x}} \cdot\left(1-\frac{1}{1+e^{-x}}\right) \\
& =\sigma(x) \cdot(1-\sigma(x))
\end{aligned}

So the gradients can be expressed with the value of the forward pass!
I think you know where this is going ; ).
